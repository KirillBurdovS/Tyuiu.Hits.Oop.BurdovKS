```
using System;


// —Å–¥–µ–ª–∞–ª, —á—Ç–æ –±—ã –≤—ã–≤–æ–¥–∏–ª–æ—Å—å
public static class Output
{
    public static void Display(object message)
    {
        #if !POLYGLOT
        Console.WriteLine(message);
        #else
        
        Console.WriteLine(message);
        #endif
    }
}


public class Product
{
    public string Name { get; set; }
    public decimal Price { get; set; }
    public string Manufacturer { get; set; }

    public Product(string name, decimal price, string manufacturer)
    {
        Name = name;
        Price = price;
        Manufacturer = manufacturer;
    }

    public virtual string GetInfo()
    {
        return $"–ù–∞–∑–≤–∞–Ω–∏–µ: {Name}, –¶–µ–Ω–∞: {Price} —Ä—É–±., –ü—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å: {Manufacturer}";
    }

    public virtual void Discount(decimal discountPercent)
    {
        if (discountPercent > 0 && discountPercent <= 100)
        {
            decimal discountAmount = Price * (discountPercent / 100);
            Price -= discountAmount;
            Output.Display($"–ü—Ä–∏–º–µ–Ω–µ–Ω–∞ —Å–∫–∏–¥–∫–∞ {discountPercent}%. –ù–æ–≤–∞—è —Ü–µ–Ω–∞: {Price:F2} —Ä—É–±.");
        }
        else
        {
            Output.Display("–ù–µ–≤–µ—Ä–Ω—ã–π –ø—Ä–æ—Ü–µ–Ω—Ç —Å–∫–∏–¥–∫–∏!");
        }
    }

    public virtual void Display()
    {
        Output.Display("=== –ò–ù–§–û–†–ú–ê–¶–ò–Ø –û –ü–†–û–î–£–ö–¢–ï ===");
        Output.Display(GetInfo());
        Output.Display("==============================");
    }
}

// –í–µ—Å—å –∫–ª–∞—Å—Å –¥–ª—è —ç–ª–µ–∫—Ç—Ä–æ–Ω–∏–∫–∏
public class Electronics : Product
{
    public int WarrantyPeriod { get; set; }

    public Electronics(string name, decimal price, string manufacturer, int warrantyPeriod) 
        : base(name, price, manufacturer)
    {
        WarrantyPeriod = warrantyPeriod;
    }

    public override void Discount(decimal discountPercent)
    {
        if (WarrantyPeriod < 12)
        {
            decimal maxDiscount = Math.Min(discountPercent, 10);
            Output.Display($"–í–Ω–∏–º–∞–Ω–∏–µ! –ì–∞—Ä–∞–Ω—Ç–∏—è –≤—Å–µ–≥–æ {WarrantyPeriod} –º–µ—Å—è—Ü–µ–≤. –ú–∞–∫—Å. —Å–∫–∏–¥–∫–∞: {maxDiscount}%");
            base.Discount(maxDiscount);
        }
        else
        {
            Output.Display($"–ì–∞—Ä–∞–Ω—Ç–∏—è {WarrantyPeriod} –º–µ—Å—è—Ü–µ–≤. –ü—Ä–∏–º–µ–Ω—è–µ–º –ø–æ–ª–Ω—É—é —Å–∫–∏–¥–∫—É!");
            base.Discount(discountPercent);
        }
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", –ì–∞—Ä–∞–Ω—Ç–∏—è: {WarrantyPeriod} –º–µ—Å—è—Ü–µ–≤";
    }

    public void CheckWarranty()
    {
        if (WarrantyPeriod > 24)
        {
            Output.Display("‚≠ê –ü—Ä–µ–º–∏–∞–ª—å–Ω–∞—è –≥–∞—Ä–∞–Ω—Ç–∏—è!");
        }
        else if (WarrantyPeriod > 12)
        {
            Output.Display("‚úÖ –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∞—è –≥–∞—Ä–∞–Ω—Ç–∏—è");
        }
        else
        {
            Output.Display("‚ö†Ô∏è –ö–æ—Ä–æ—Ç–∫–∞—è –≥–∞—Ä–∞–Ω—Ç–∏—è");
        }
    }
}

// –í–µ—Å—å –∫–ª–∞—Å—Å –¥–ª—è –æ–¥–µ–∂–¥—ã
public class Clothing : Product
{
    public string Size { get; set; }

    public Clothing(string name, decimal price, string manufacturer, string size) 
        : base(name, price, manufacturer)
    {
        Size = size;
    }

    public override void Display()
    {
        Output.Display("=== –û–î–ï–ñ–î–ê ===");
        Output.Display($"–ù–∞–∑–≤–∞–Ω–∏–µ: {Name}");
        Output.Display($"–¶–µ–Ω–∞: {Price} —Ä—É–±.");
        Output.Display($"–ü—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—å: {Manufacturer}");
        Output.Display($"–†–∞–∑–º–µ—Ä: {Size}");
        Output.Display("==============");
    }

    public override string GetInfo()
    {
        return base.GetInfo() + $", –†–∞–∑–º–µ—Ä: {Size}";
    }

    public void TryOn()
    {
        Output.Display($"–ü—Ä–∏–º–µ—Ä–∫–∞ –æ–¥–µ–∂–¥—ã '{Name}' —Ä–∞–∑–º–µ—Ä–∞ {Size}...");
        Output.Display("–ü—Ä–∏–º–µ—Ä–∫–∞ –∑–∞–≤–µ—Ä—à–µ–Ω–∞ —É—Å–ø–µ—à–Ω–æ!");
    }
}

// –í–µ—Å—å –∫–ª–∞—Å—Å –¥–ª—è –∫–Ω–∏–≥
public class Books : Product
{
    public string Author { get; set; }

    public Books(string name, decimal price, string manufacturer, string author) 
        : base(name, price, manufacturer)
    {
        Author = author;
    }

    public override string GetInfo()
    {
        return $"–ö–Ω–∏–≥–∞: {Name}, –ê–≤—Ç–æ—Ä: {Author}, –¶–µ–Ω–∞: {Price} —Ä—É–±., –ò–∑–¥–∞—Ç–µ–ª—å—Å—Ç–≤–æ: {Manufacturer}";
    }

    public override void Discount(decimal discountPercent)
    {
        decimal actualDiscount = Math.Min(discountPercent, 30);
        Output.Display($"–î–ª—è –∫–Ω–∏–≥ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è —Å–∫–∏–¥–∫–∞ 30%. –ü—Ä–∏–º–µ–Ω—è–µ–º: {actualDiscount}%");
        base.Discount(actualDiscount);
    }

    public void ReadSample()
    {
        Output.Display($"–ß—Ç–µ–Ω–∏–µ –æ—Ç—Ä—ã–≤–∫–∞ –∏–∑ –∫–Ω–∏–≥–∏ '{Name}' –∞–≤—Ç–æ—Ä–∞ {Author}...");
        Output.Display("¬´–≠—Ç–æ –±—ã–ª–∞ —Ç–µ–º–Ω–∞—è –∏ stormy night...¬ª");
    }
}

// —Å–æ–∑–¥–∞–ª –æ–±—ä–µ–∫—Ç—ã, —á—Ç–æ –±—ã –ø–æ—Ä–∞–±–æ—Ç–∞—Ç—å —Å –Ω–∏–º–∏
var laptop = new Electronics("–ù–æ—É—Ç–±—É–∫ Gaming", 50000, "ASUS", 24);
var shirt = new Clothing("–•–ª–æ–ø–∫–æ–≤–∞—è —Ä—É–±–∞—à–∫–∞", 2500, "Zara", "L");
var novel = new Books("–í–æ–π–Ω–∞ –∏ –º–∏—Ä", 1500, "–≠–∫—Å–º–æ", "–õ–µ–≤ –¢–æ–ª—Å—Ç–æ–π");

Output.Display("üõçÔ∏è –°–ò–°–¢–ï–ú–ê –£–ü–†–ê–í–õ–ï–ù–ò–Ø –ü–†–û–î–£–ö–¢–ê–ú–ò");
Output.Display("==================================");
Output.Display("");

// 1. –¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ —ç–ª–µ–∫—Ç—Ä–æ–Ω–∏–∫–∏
Output.Display("1. –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–ï –≠–õ–ï–ö–¢–†–û–ù–ò–ö–ò");
laptop.Display();
laptop.CheckWarranty();
laptop.Discount(15);
Output.Display("");

// 2. –¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –æ–¥–µ–∂–¥—ã
Output.Display("2. –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–ï –û–î–ï–ñ–î–´");
shirt.Display();
shirt.TryOn();
shirt.Discount(20);
Output.Display("");

// 3. –¢–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ –∫–Ω–∏–≥
Output.Display("3. –¢–ï–°–¢–ò–†–û–í–ê–ù–ò–ï –ö–ù–ò–ì");
novel.Display();
Output.Display("–ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è: " + novel.GetInfo());
novel.ReadSample();
novel.Discount(40);
Output.Display("");

// 4. –ü–æ–ª–∏–º–æ—Ä—Ñ–∏–∑–º–∞
Output.Display("4. –î–ï–ú–û–ù–°–¢–†–ê–¶–ò–Ø –ü–û–õ–ò–ú–û–†–§–ò–ó–ú–ê");
Product[] products = {
    new Electronics("–°–º–∞—Ä—Ç—Ñ–æ–Ω", 30000, "Samsung", 6),
    new Clothing("–î–∂–∏–Ω—Å—ã", 4000, "Levi's", "M"),
    new Books("1984", 800, "–ê–°–¢", "–î–∂–æ—Ä–¥–∂ –û—Ä—É—ç–ª–ª")
};

foreach (var product in products)
{
    product.Display();
    product.Discount(25);
    Output.Display("");
}

Output.Display("–ü–†–û–ì–†–ê–ú–ú–ê –£–°–ü–ï–®–ù–û –ó–ê–í–ï–†–®–ï–ù–ê!");
```
